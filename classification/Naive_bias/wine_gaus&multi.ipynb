```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
model_gaus=GaussianNB()
model_multi=MultinomialNB()
from sklearn.model_selection import cross_val_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(wine.data,wine.target,train_size=0.8)
model_gaus.fit(x_train,y_train)
```

```python
model_multi.fit(x_train,y_train)
```

```python
gaus_scores=cross_val_score(model_gaus,x_test,y_test)
multi_scores=cross_val_score(model_multi,x_test,y_test)
```

```python
import numpy as np
```

```python
print(np.mean(gaus_scores))
```

```python
print(np.mean(multi_scores))
```

```python
model_gaus.score(x_test,y_test)
```

```python
model_multi.score(x_test,y_test)
```
