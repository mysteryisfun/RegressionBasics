```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le_gender=LabelEncoder()
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

data sets need are target:survived
input: class,age, sex, fare

```python
df['gender']=le_gender.fit_transform(df['Sex'])
df.head()
```

```python
x=df.drop(['PassengerId','Survived','Name','Sex','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
y=df.Survived
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model=tree.DecisionTreeClassifier(criterion="entropy")
model.fit(x_train,y_train)
model.score(x_test,y_test)
#entropy has still lower score
```
