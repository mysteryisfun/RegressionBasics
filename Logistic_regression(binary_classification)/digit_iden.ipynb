```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Age': [22, 35, 45, 31, 50, 23, 40, 28],
    'Income': ['Low', 'Medium', 'High', 'Medium', 'High', 'Low', 'Medium', 'Low'],
    'BuysProduct': ['No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}

# Creating DataFrame
df = pd.DataFrame(data)

# Encoding categorical variables (Income and BuysProduct)
df['Income'] = df['Income'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['BuysProduct'] = df['BuysProduct'].map({'No': 0, 'Yes': 1})

# Selecting features (Age and Income) and target (BuysProduct)
X = df[['Age', 'Income']]
y = df['BuysProduct']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Visualize the decision tree
plt.figure(figsize=(10, 8))
tree.plot_tree(clf, filled=True, feature_names=['Age', 'Income'], class_names=['No', 'Yes'], rounded=True)
plt.show()
```
